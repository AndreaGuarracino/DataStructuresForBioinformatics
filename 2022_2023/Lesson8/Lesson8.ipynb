### Interfaces

Matplotlib provides two interfaces, which could be confusing.

#### `pyplot` interface

The interface with which Matplotlib was born.

```python
plt.figure()

plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, y)

plt.subplot(2, 1, 2)
plt.plot(x, np.log(x));
```

#### Object-oriented interface

In the object-oriented interface the plotting functions are methods of `Figure` and `Axes` objects.

```python
fig, ax = plt.subplots(2, 1)

ax[0].plot(x, y)
ax[1].plot(x, np.log(x));
```

```python
fig, ax = plt.subplots(2, 1)
```

#### Line colors

```python
plt.figure()

plt.subplot(2, 1, 1)
plt.plot(x, y, color='red')                # color name 

plt.subplot(2, 1, 2)
plt.plot(x, np.log(x), color='#FFAA00');   # hex code (RGB)
```

```python
fig, ax = plt.subplots(2, 1)

ax[0].plot(x, y, color='red')              # color name
ax[1].plot(x, np.log(x), color='#FFAA00'); # hex code (RGB)
```

#### Line styles

```python
plt.figure()

plt.subplot(2, 1, 1)
plt.plot(x, y, linestyle='-')             # solid

plt.subplot(2, 1, 2)
plt.plot(x, np.log(x), linestyle='--');   # dashed
```

```python
fig, ax = plt.subplots(2, 1)

ax[0].plot(x, y, linestyle='-')           # solid
ax[1].plot(x, np.log(x), linestyle='--'); # dashed
```

#### Axes limits

```python
plt.figure()

plt.subplot(2, 1, 1)
plt.plot(x, y)
plt.xlim(-5, 10)

plt.subplot(2, 1, 2)
plt.plot(x, np.log(x))
plt.ylim(0, 20);
```

```python
fig, ax = plt.subplots(2, 1)

ax[0].plot(x, y)
ax[1].plot(x, np.log(x))

ax[0].set_xlim(-5, 10)
ax[1].set_ylim(0, 20);
```

#### Labels

```python
plt.figure()

plt.subplot(2, 1, 1)
plt.plot(x, y)
plt.ylabel('y A')
plt.title('Main title')

plt.subplot(2, 1, 2)
plt.plot(x, np.log(x))
plt.ylabel('y B')

plt.xlabel('x');
```

```python
fig, ax = plt.subplots(2, 1)

ax[0].plot(x, y)
ax[1].plot(x, np.log(x))

ax[0].set_ylabel('y A')

ax[1].set_xlabel('x')
ax[1].set_ylabel('y B')

ax[0].set_title('Main title');
```

#### Saving Figure on File

```python
fig, ax = plt.subplots(2, 1)

ax[0].plot(x, y)
ax[1].plot(x, np.log(x))

ax[0].set_ylabel('y A')

ax[1].set_xlabel('x')
ax[1].set_ylabel('y B')

ax[0].set_title('Main title')

fig.savefig('myPlots.pdf');
```

### \*args and \**kwargs

#### \*args
*args is used to send a non-keyworded variable length argument list to the function. 

```python
def function_argv(first_argument, *args):
    print("first_argument:", first_argument)
    
    for i, arg in enumerate(args):
        print("other_arguments (n. {}): {}".format(i + 1, arg))
```

```python
function_argv('Protein', 'PDB_ID1', 'PDB_ID2', 'PDB_ID3')
```

```python
function_argv('Protein')
```

```python
args_t = ("two", 3, 5)

function_argv(*args_t)
```

#### \**kwargs

**kwargs allows you to pass keyworded variable length of arguments to a function. 

```python
def function_kwargs(first_argument, **kwargs):
    print('first_argument:', first_argument)
    
    for i, (key, value) in enumerate(kwargs.items()):
        print("other_arguments (n. {}): {} - {}".format(i + 1, key, value))
```

```python
function_kwargs('zero', first="1", second="2", third="3")
```

```python
function_kwargs('zero')
```

```python
kwargs = {"arg3": 3, "arg2": "two", "arg1": 5}

function_kwargs("zero", **kwargs)
```
